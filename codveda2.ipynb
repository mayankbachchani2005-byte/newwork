```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset loaded
df=pd.read_csv(r"D:\python work\cleaned_data_of_stock_price.csv")
```

```python
print(df.head())
```

```python
# summary stas
print("mean is:\n",df.mean(numeric_only=True))
```

```python
print("median is:\n",df.median(numeric_only=True))
```

```python
print("mode",df.mode())
```

```python
print("standard deviations :\n",df.std(numeric_only=True))
```

```python
# describe
print(df.describe())
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.show()
```

```python
plt.scatter(df["open"],df["close"])
plt.xlabel("open price")
plt.ylabel("close price")
plt.title("open price vs close price")
plt.show()
```

```python
correlations=df.corr(numeric_only=True)
print(correlations)
```

```python
plt.Figure(figsize=(8,6))
sns.heatmap(correlations,annot=True,cmap="coolwarm")
plt.show()
```
