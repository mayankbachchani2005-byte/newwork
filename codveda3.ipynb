```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv(r"D:\python work\cleaned_data_of_stock_price.csv")
print(dataset.head(3))
```

```python
# line chart
plt.Figure(figsize=(8,5))

plt.plot(dataset["open"],label="open price")
plt.plot(dataset["close"],label="close price")

plt.title("stock price trends")
plt.xlabel("index")
plt.ylabel("price")

plt.show()
```

```python
# bar chart
avg_price=dataset[["open","high","low","close"]].mean()

avg_price.plot(kind="bar")
plt.title("avg_stock_price")
plt.xlabel("price type")
plt.ylabel("average value")

plt.show()
```

```python
plt.scatter(dataset["open"],dataset["close"])

plt.title("open vs close price")
plt.xlabel("open price")
plt.ylabel("close price")

plt.show()
```
